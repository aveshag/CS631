```python
%matplotlib notebook
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data=pd.read_excel("HW_TESLA.xlt")        #reading data
feature_name=list(data.columns.values)
data=data.values
X = data[:,1:]                    #separating labels from features
y = data[:,0]

print(X.shape)
```

```python

kernel = metrics.pairwise.rbf_kernel(X, Y=None, gamma=None)
D = kernel.sum(axis=1)
D_inv = D**(-1/2)

D_inv_mat = np.diag(D_inv)

temp = np.dot(D_inv_mat, kernel) 
P_d = np.dot(kernel, D_inv_mat) 

print(P_d)
print(P_d.shape)
```

```python
w , v = np.linalg.eigh(P_d)
```

```python
print(w)
X_ = v*(w.T)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, max_iter = 300) 
kmeans.fit(X_)

# KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
#     n_clusters=2, n_init=10, n_jobs=1, precompute_distances='auto',
#     random_state=None, tol=0.0001, verbose=0)
```

```python
correct = 0
zero=0
one=0
for i in range(len(X_)):
    predict_me = np.array(X_[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = kmeans.predict(predict_me)
    prediction[0] = 1 - prediction[0]
    if (prediction[0]) == 0:
        zero += 1
    if (prediction[0]) == 1:
        one += 1
    if prediction[0] == y[i]:
        correct += 1

print(correct/len(X_))
print(zero)
print(one)
```

```python
X_3 = X_[:,-3:]
print(X_3.shape)
```

```python
kmeans.fit(X_3)
```

```python
correct = 0
zero = 0
one = 0
for i in range(len(X_3)):
    predict_me = np.array(X_3[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = kmeans.predict(predict_me)
    #prediction[0] = 1 - prediction[0]
    if (prediction[0]) == 0:
        zero += 1
    if (prediction[0]) == 1:
        one += 1
    if prediction[0] == y[i]:
        correct += 1

print(correct/len(X_))
print(zero)
print(one)
```

```python
fig1=plt.figure()
ax1=fig1.add_subplot(111,projection='3d')
ax1.scatter(X_3[:,0],X_3[:,1],X_3[:,2], marker='s',c=y ,cmap='RdBu')
fig2=plt.figure()
ax2=fig2.add_subplot(111,projection='3d')
ax2.scatter(X_3[:,0],X_3[:,1],X_3[:,2], marker='s',c=kmeans.labels_, cmap='RdBu')
fig2.show()
```
